## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

We would predict whether a page matches an observed pattern that makes it worth reviewing for a content-refresh opportunity. The target label would come from an observed outcome in the dataset, such as a page’s subsequent performance or trend direction. Hence, since the dataset does not observe whether a page was refreshed and whether that refresh caused improvement, the target is only a proxy for “worth reviewing.” The model would therefore learn to rank pages based on how much they resemble pages associated with an observed outcome, not predict with certainty that refreshing a particular page will cause its performance to improve.
